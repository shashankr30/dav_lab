```python
n = int(input("Enter the number of elements to add: "))
l =[]
for i in range(n):
    l.append(int(input("Enter the element: ")))
num = int(input("Enter number to search: "))

print("Smallest: ",sorted(l)[0])
print("Largest: ", sorted(l)[-1])
print("Sum: ",sum(l)) 
print("Average: ",sum(l)/len(l))
print(f'Occurrences of {num} is: ',l.count(num))    
```

```python
from collections import Counter

wc_win = {}
n = int(input("Enter the number of entries: "))
for i in range(n):
    wc_win[int(input("Enter the year: "))] = input("Enter the country: ")
x = list(wc_win.values())
suc = Counter(wc_win.values())
for i in x:
    if x.count(i)>1:
        x.remove(i)
print("The best performing country is: ", suc.most_common(1)[0][0])
print("All the winners: ")
for i in x:
    print(i, end =" ")
    
```

```python
x = input("Enter a sentence: ")
wrd = []
dict_wrd = {}
for i in x.split():
    wrd.append(i)
for i in wrd:
    dict_wrd[str(i)] = x.count(i)
print(dict_wrd)
```

```python
n1 = int(input("Enter the number of elements in set 1: "))
n2 = int(input("Enter the number of elements in set 2: "))
s1=set()
s2=set()
for i in range(n1):
    s1.add(int(input("Enter element for set 1: ")))
for i in range(n2):
    s2.add(int(input("Enter element for set 2: ")))
print("Union: ", s1.union(s2))
print("Intersection: ", s1.intersection(s2))
print("Difference: ", s1.difference(s2))
```
