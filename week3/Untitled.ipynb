```python
deg = [35,40,32,33,29,34,37]
temp  = pd.Series(deg, index = ["Monday", "Tuesday","Wednesday", "Thursday","Friday","Saturday","Sunday"])
print("Mean is: ",temp.mean())
print("Max is: ",temp.max(), "on",temp.idxmax())
print("Min is: ",temp.min(),"on",temp.idxmin())
v = int(input("Enter the parameter value: "))
print(temp[temp>v])
f = temp*(9/5)+32
print("Temp in fahrenheit:\n", f)
print("Days greater than Average:\n",temp[temp>temp.mean()])
```

```python
data = {
    "Roll Number": range(1, 11),
    "Name": ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J"],
    "Gender": ["M","F","M","F","M","F","M","F","M","F"],
    "Marks1": [45, 67, 89, 34, 56, 78, 90, 23, 49, 65],
    "Marks2": [55, 38, 92, 40, 76, 59, 33, 81, 47, 69],
    "Marks3": [72, 64, 88, 53, 91, 77, 85, 60, 70, 82]
}
 
df = pd.DataFrame(data)
df["Total"] = df["Marks1"] + df["Marks2"] + df["Marks3"]
print("Lowest Marks1:", df["Marks1"].min())
print("Highest Marks2:", df["Marks2"].max())
print("Average Marks3:", df["Marks3"].mean())
df["Average"] = df[["Marks1","Marks2","Marks3"]].mean(axis=1)
top_student = df.loc[df["Average"].idxmax(), "Name"]
print("Top Student:", top_student)

failed_count = (df["Marks2"] < 40).sum()
print("Number of students failed in Marks2:", failed_count)
```
