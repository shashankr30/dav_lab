```python
import numpy as np

arr = np.random.randint(1, 51,12).reshape(3,4)
print(arr)
print("Mean: ", np.mean(arr))
print("Median: ",np.median(arr))
print("Standard Deviation: ",np.std(arr))
print("Total Sum", np.sum(arr))
print("Row Sum", np.sum(arr, axis =1))
print("Reshaped: ",arr.reshape((2,6)))
```

```python
import time

# Record the start time
start_time = time.time()

x = np.arange(1,10).reshape(3,3)
y = np.arange(10,19).reshape(3,3)
print(f"Sum: {x+y}\nDiff: {x-y}\nDot-Prod: {np.dot(x,y)}\nElement-Multiply: {x*y}\nDivision: {x/y}\nTranpose M1: {x.T}\nTranspose M2: {y.T}" )
print(f"Det of m1:{np.linalg.det(y)}, Det of m2: {np.linalg.det(x)}")
if np.linalg.det(x) != 0: 
    print("Inverse of m1:\n", np.linalg.inv(x)) 
else: 
    print("Matrix 1 is not invertible")
if np.linalg.det(y) != 0: 
    print("Inverse of m2:\n", np.linalg.inv(y)) 
else: 
    print("Matrix 2 is not invertible")
elapsed_time = time.time() - start_time
print(elapsed_time)
```
