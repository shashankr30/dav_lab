```python
#WAP to check the largest among the given three numbers

def largest_three(a,b,c):
    if a>b:
        if a>c:
            print(a," is the largest.")
        else:
            print(c, " is the largest.")
    elif b>c:
        print(b, "is the largest.")
    else:
        print(c, "is the largest.")

a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
c = int(input("Enter the third number: "))
largest_three(a,b,c)
```

```python
def prime_check(x):
    prime = True
    for i in range(2,x):
        if x%i==0:
            prime = False
    if x ==1:
        prime = "Neither prime nor composite"
    return prime
a = int(input("Enter a number: "))
prime_check(a)
```

```python
def count_letters():
    v = ['a','e','i','o','u']
    x = input("Enter a string: ").lower().split()
    nv=nc=0
    for i in x:
        for j in i:
            if j in v:
                nv+=1
            else:
                nc+=1
    print("Consonants: ",nc)
    print("Vowels: ",nv)   
count_letters()
```
